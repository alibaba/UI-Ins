```python
random50k_df = pd.read_json("/braincoder-extreme-nas/panrong/data/gta1/random50k/random50k.json")
image_list = random50k_df["image"].to_list()
with open("/braincoder-extreme-nas/panrong/data/gta1/random50k/image_list.txt", "w") as f:
    for image in image_list:
        f.write(image + "\n")
```

```python
import json
from sklearn.model_selection import train_test_split

with open("/braincoder-extreme-nas/panrong/data/gta1/random50k/random50k.json", "r") as f:
    data_all = json.load(f)

# 将data_all划分成训练集和验证集，验证集占10%
data_train, data_val = train_test_split(data_all, test_size=0.1, random_state=42, shuffle=True)

# 分别存储为json文件
with open("/braincoder-extreme-nas/panrong/data/gta1/random50k/random50k_train.json", "w") as f:
    json.dump(data_train, f)
with open("/braincoder-extreme-nas/panrong/data/gta1/random50k/random50k_val.json", "w") as f:
    json.dump(data_val, f)
```
